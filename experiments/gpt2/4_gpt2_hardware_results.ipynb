# GPT2 Results Analysis

## Hardware juicing results

Here we look at the effect of different harware on tokens per second (TPS) during training.

```python
import matplotlib.pyplot as plt
import pandas as pd


hw_df = pd.DataFrame(
    {
        "hardware": ["cpu", "mps", "RTX3090", "H100"],
        "tps": [645, 1499, 11381, 27026],
        "mfu": [pd.NA, pd.NA, 27.01, 34.47],
    }
)
# add column for speedup over "cpu"
hw_df["speedup"] = hw_df["tps"] / hw_df["tps"].iloc[0]
hw_df
```

```python
# Bar plot of TPS for each harware
x = hw_df["hardware"]
y = hw_df["tps"]

plt.bar(x, y)
plt.xlabel("Hardware")
plt.ylabel("Tokens per second")
plt.title("Tokens per second for each hardware")
plt.show()
```

```python
# Bar plot of Speedup for each hardware
x = hw_df["hardware"]
y = hw_df["speedup"]

plt.bar(x, y)
plt.xlabel("Hardware")
plt.ylabel("X Speedup over CPU")
```
